```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(15, 5))

# Add x-axis and y-axis
ax.bar(df.loc["2007-01-01":"2010-12-31"].index.values,
       df.loc["2007-01-01":"2010-12-31", "Close"],
       color='green')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Closing Prices ($)",
       title="Prices for AMZN Stock\n 2007 - 2010")

# plot vertical lines at two dates from y=0 to y=250
ax.vlines(x=['2008-01-01', '2009-12-31'], ymin=0, ymax=df.loc["2007-01-01":"2010-12-31", "Close"].max(), color='r', ls='--')

# Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(15, 5))

# Add x-axis and y-axis
ax.bar(df.index.values,
       df['Close'],
       color='green')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Closing Prices ($)",
       title="Prices for AMZN Stock\n 1997 - 2022")

# Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```
