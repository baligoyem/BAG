```python
dt_parser = lambda x: datetime.strptime(x, '%Y %m %d %H')
df = pd.read_csv(r"C:\Users\beytu\BAG_notebooks\Datasets\air-quality-india.csv",
                 usecols=["Year", "Month", "Day", "Hour", "PM2.5"],
                 parse_dates={"Date": ["Year", "Month", "Day", "Hour"]},
                 date_parser=dt_parser)
df.tail()
```

```python
dt_converter = lambda x: pd.to_datetime(x, dayfirst=True)

df = pd.read_csv(r"C:\Users\beytu\BAG_notebooks\Datasets\AMZN.csv", converters={"Date":dt_converter})   
# to convert values in the columns by defined functions.
df.tail()
```

```python
survey_df = pd.read_csv('videogames_survey.csv',
                        true_values=['Yes', 'YES', 'yes'],
                        false_values=['no', 'No', 'NO'],
                        na_values=['?', '-'],
                        nrows=10)
# na_values: np.NaN value is replaced with invalid values
# nrows: first n rows of data file are imported as a sample
survey_df
```
